```python
 #a,b,c,d,e denoting the number of times there is a battery level in each respective range 
```

```python
def getBPercentageList(battery):
    global a
    global b
    global c
    global d
    global e
    count = 0 

    length = len(battery)


    while count < length:
        if (battery[count] == 0):
            
            a = a + 1
        elif (battery[count] < 26):
            
            b = b + 1
        elif (battery[count] < 51):
            
            c = c + 1
        elif (battery[count] < 76):
            
            d = d + 1
        else:
           
            e = e + 1
        count = count + 1
    
    return [a,b,c,d,e]
    
    
```

```python
def createChart(val):
    label = ["0%", "1%-25%", "26%-50%", "51%-75%", "76%-100%"]
    values = val

    colors = ['k', 'r', 'y', 'b', 'g']
    plt.pie(values, labels=label, colors=colors, startangle=90, autopct='%.1f%%')
    plt.show()

# trace = go.Pie(labels=labels, values=values)

# py.iplot([trace], filename='basic_pie_chart')
```
